# cos wave

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t=np.arange(0.0,2.0,0.01)
s=1+np.cos(2*np.pi*t)
plt.plot(t,s)
plt.xlabel("Time(x)")
plt.ylabel("Voltage(mv)")
plt.title("Cosine wave plot(Cos(x))")
plt.show()
```

```python
plt.plot(t,s,'+')
plt.xlabel("Time(x)")
plt.ylabel("Voltage(mv)")
plt.title("Cosine wave plot(Cos(x))")
plt.show()
```

# Subplot

```python
plt.plot(t,s,'--')
plt.xlabel("Time(x)")
plt.ylabel("Voltage(mv)")
plt.title("Cosine wave plot(Cos(x))")
plt.show()
```

```python
x1=np.linspace(0.5,5.0)
x2=np.linspace(0.0,2.0)
y1=np.cos(2*np.pi*x1)*np.exp(-x1)
y2=np.cos(2*np.pi*x2)
plt.subplot(2,1,1)
plt.plot(x1,y1,'o-')
plt.title("Sublot-1")
plt.xlabel("x1")
plt.ylabel("Amp(y1)")

plt.subplot(2,1,2)
plt.plot(x2,y2,'o-')
plt.title("Sublot-2")
plt.xlabel("x2")
plt.ylabel("Amp(y2)")
plt.show()
```

# Bar graph

```python
x=[1,2,3,4,5]
y=[10,45,22,55,67]
tick_labels=['one','two','three','four','five']
plt.bar(x,y,tick_label=tick_labels,width=0.5,color=["green","yellow","red"])
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

# Scatter plot

# histogram

```python
np.random.seed(10**7) 
mu = 121 
sigma = 21
x = mu + sigma * np.random.randn(1000) 
   
num_bins = 100
   
n, bins, patches = plt.hist(x, num_bins,  
                            density = 1,  
                            color ='green', 
                            alpha = 0.7) 
   
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2)) 
  
plt.plot(bins, y, '--', color ='black') 
  
plt.xlabel('X-Axis') 
plt.ylabel('Y-Axis') 
  
plt.title('matplotlib.pyplot.hist() function Example\n\n', 
          fontweight ="bold") 
  
plt.show() 
```
