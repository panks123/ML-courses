```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
import seaborn as sns
import math
%matplotlib inline
```

```python
data=pd.read_csv('titanic.csv')
data
```

```python
sns.heatmap(data.isnull(),cmap="Greens")
```

```python
data.drop('cabin',axis=1,inplace=True)
```

```python
data.columns
```

```python
pd.get_dummies(data['sex'])
```

```python
sex=pd.get_dummies(data['sex'],drop_first=True)
sex.head()
```

```python
pd.get_dummies(data['embarked'])
```

```python
embark=pd.get_dummies(data['embarked'],drop_first=True)
embark
```

```python
pclass=pd.get_dummies(data['pclass'],drop_first=True)
pclass
```

```python
data=pd.concat([data,sex,embark,pclass],axis=1)
data.head()
```

```python
data.drop(['sex','name','ticket','embarked'],axis=1,inplace=True)
```

```python
data.drop('pclass',axis=1,inplace=True)
```

```python
x=data.drop('survived',axis=1)
x=x.drop('home.dest',axis=1)
x=x.drop('boat',axis=1)
x=x.drop('body',axis=1)
x.age.fillna(28,inplace=True)
x.sibsp.fillna(1,inplace=True)
x.parch.fillna(0,inplace=True)
x.fare.fillna(x.fare.mean(),inplace=True)
y=data['survived']
y.fillna(1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression as LR
logmodel=LR()
logmodel.fit(x_train,y_train)
```

```python
prediction=logmodel.predict(x_test)
prediction
```

```python
pd.DataFrame(prediction,y_test)
```
