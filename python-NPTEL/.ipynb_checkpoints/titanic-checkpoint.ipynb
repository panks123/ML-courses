```python
df1.pclass.fillna(3,inplace=True)
df1.survived.fillna(0,inplace=True)
df1.age.fillna(30,inplace=True)
df1.sibsp.fillna(0,inplace=True)
df1.parch.fillna(0,inplace=True)
df1.fare.fillna(np.mean(df1.fare),inplace=True)
df1.boat.fillna("NO boat",inplace=True)
df1["home.dest"].fillna("New York",inplace=True)
df1.embarked.fillna("S",inplace=True)
df1.name.fillna("Miss Kate",inplace=True)
df1.ticket.fillna(np.random.randint(1000,2000),inplace=True)
df1.sex.fillna("male",inplace=True)
df1.isnull().sum()
```

```python
sns.countplot(x="survived",data=df1)
```

```python
sns.countplot(x="survived",hue="sex",data=df1)
```

```python
z=[]
for ab in list(df1['age']):
    if ab<=18:
        z.append('c')
    elif ab<=50 and ab>18:
        z.append('a')
    else:
        z.append('o')
df1['agegroup']=z
        
```

```python
w=[]
for ab in list(df1['age']):
    if ab<=15 and ab<20:
        w.append("teen")
    elif ab<=50 and ab>18:
        w.append("twenties")
    else:
        w.append('other')
df1['teens and twenties']=w
df1        
```

```python
sns.countplot(x='sex', hue="teens and twenties",data=df1)
```
