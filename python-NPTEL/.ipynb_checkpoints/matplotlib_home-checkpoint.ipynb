```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.style.use("seaborn-whitegrid")
x=np.linspace(0,10,100)
fig=plt.figure()
plt.plot(x,np.sin(x),'-')
plt.plot(x,np.cos(x),'--')
```

```python
plt.plot(x,np.tan(x),':')
```

```python
plt.figure()
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))
plt.subplot(2,1,2)
plt.plot(x,np.cos(x))
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x))
```

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

```python
plt.plot(x,np.sin(x),color='blue')
plt.plot(x,np.sin(x+1),color='r')
plt.plot(x,np.sin(x+2),color='#000000')
plt.plot(x,np.sin(x+3),color=(1,.2,.3))
plt.plot(x,np.sin(x+4),color='chartreuse')
plt.plot(x,np.sin(x),color='.39')
```

```python


plt.plot(x,x+2,linestyle='solid')
plt.plot(x,x+3,linestyle='dashed')
plt.plot(x,x+4,linestyle='dashdot')
plt.plot(x,x+1,linestyle='dotted')
plt.show()

plt.plot(x,x+2,linestyle='-')
plt.plot(x,x+3,linestyle='--')
plt.plot(x,x+4,linestyle='-.')
plt.plot(x,x+1,linestyle=':')
```

```python
plt.plot(x,x+2,'-g')
plt.plot(x,x+3,'--k')
plt.plot(x,x+4,'-.r')
plt.plot(x,x+1,':y')
```

```python
plt.plot(x,np.sin(x))
plt.xlim(-1,13)
plt.ylim(-1.5,1.5)
```

```python
plt.plot(x,np.sin(x))
plt.axis('tight')
```

```python
plt.plot(x,np.sin(x))
plt.axis('equal')
```

```python
plt.plot(x,np.sin(x))
plt.title('Curve : f(x)=sin(x)')
plt.xlabel('x')
plt.ylabel('y')
```
