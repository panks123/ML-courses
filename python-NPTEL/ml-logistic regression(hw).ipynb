```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('advertising.csv')
data
```

```python
sns.pairplot(data=data,hue='Clicked on Ad',palette='coolwarm')
```

```python
sns.distplot(data['Age'],bins=40,kde=True)
```

```python
sns.heatmap(data.isnull(),cmap="Greens")
```

```python
data.drop(['Ad Topic Line','City','Country','Timestamp'],axis=1,inplace=True)
data.head(10)
```

```python
y=data['Clicked on Ad'].values
y
```

```python
x=data.drop('Clicked on Ad',axis=1)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.33,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
x_train=sc_x.fit_transform(x_train)
x_test=sc_x.transform(x_test)
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression as LR
classifier=LR(random_state=0)
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)*100
accuracy
```
