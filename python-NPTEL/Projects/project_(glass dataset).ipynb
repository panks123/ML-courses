```python
def normalise(glassdata):
    result=glassdata.copy()
    for feature_name in glassdata.columns:
        max_value =glassdata[feature_name].max()
        min_value=glassdata[feature_name].min()
        result[feature_name]= (glassdata[feature_name]-min_value) / (max_value - min_value)
    return result
```

```python
label=glass['Type']
```

```python
del glass['Type']
glass.head(1)
```

```python
glass=normalise(glass)
glass.head()
```

```python
y=glass['Type']
del glass['Type']
X=glass
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Applying decision tree

```python
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 1)
classifier.fit(X_train, y_train)
```

```python
pred = classifier.predict(X_test)
pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, pred)
cm
```

```python
dec_tree_ac_score=accuracy_score(y_test,pred)*100
dec_tree_ac_score
```

# Appling random forest

```python
classifier = RandomForestClassifier(n_estimators = 1000,n_jobs=10000, criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
prediction=classifier.predict(X_test)
```

```python
prediction
```

```python
from sklearn.metrics import confusion_matrix
cmatrix = confusion_matrix(y_test, prediction)
cmatrix
```

```python
rand_forest_ac_score=accuracy_score(y_test,prediction)*100
rand_forest_ac_score
```

## Applyng knn

```python
classifier = KNeighborsClassifier(n_neighbors = 7,p=2)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
c_m = confusion_matrix(y_test, y_pred)
c_m
```

```python
knn_ac_score=accuracy_score(y_test,y_pred)*100
knn_ac_score
```
