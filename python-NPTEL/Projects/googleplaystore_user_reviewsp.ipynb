```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
 
%matplotlib inline
```

```python
play_store_reviews = pd.read_csv('googleplaystore_user_reviews.csv')
df = pd.DataFrame(play_store_reviews)
df
```

```python
df.dtypes
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Translated_Review'].fillna(value='no_review', inplace=True)
df.dropna(how='any', inplace=True)
df
```

```python
df.shape
```

```python
temp = pd.get_dummies(df.Sentiment, prefix= 'Sentiment', drop_first=True)
df= pd.concat([df , temp], axis=1)
df
```

```python
plt.figure(figsize = (7,7))
df.Sentiment.value_counts().plot.pie(fontsize=14)
plt.title('All App', fontsize= 15)
```

```python
sentiment = []

for value in df.Sentiment_Polarity:
    if value<-0.8:
        sentiment.append(-5)
    elif value<-0.6:
        sentiment.append(-4)
    elif value<-0.4:
        sentiment.append(-3)
    elif value<-0.2:
        sentiment.append(-2)
    elif value<0:
        sentiment.append(-1)
    elif value==0:
        sentiment.append(0)
    elif value<=0.2:
        sentiment.append(1)
    elif value<=0.4:
        sentiment.append(2)
    elif value<=0.6:
        sentiment.append(3)
    elif value<=0.8:
        sentiment.append(4)
    elif value<=1:
        sentiment.append(5)

df['Sentiment_Rating']= sentiment
```

```python
plt.figure(figsize=(10,5))
plt.title("Categorise Sentiment Polarity of App", fontsize=15)
sns.countplot(df['Sentiment_Rating'])
plt.show()
```

```python
sentiment = []

for value in df.Sentiment_Subjectivity:
    if value==0:
        sentiment.append(0)
    elif value<=0.2:
        sentiment.append(1)
    elif value<=0.4:
        sentiment.append(2)
    elif value<=0.6:
        sentiment.append(3)
    elif value<=0.8:
        sentiment.append(4)
    elif value<=1:
        sentiment.append(5)

df['Sentiment_Sub_Categ']= sentiment
```

```python
plt.figure(figsize=(10,5))
plt.title("Categorise Sentiment Subjectivity of App", fontsize=15)
sns.countplot(df['Sentiment_Sub_Categ'])
plt.show()
```

```python
X=pd.concat([df['Sentiment_Polarity'], df['Sentiment_Subjectivity']], axis=1)
Y1=df['Sentiment_Positive']
Y2=df['Sentiment_Neutral']
```

# Applying LOgistic Regression

```python
from sklearn.model_selection import train_test_split
X_train, X_test,Y1_train,Y1_test, Y2_train,Y2_test = train_test_split(X, Y1,Y2, test_size=0.25, random_state=72)
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
regr = LogisticRegression()
regr.fit(X_train, Y1_train)
Y1_pred = regr.predict(X_test)
regr.fit(X_train, Y2_train)
Y2_pred = regr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
print("Confusion Matrix of positive Sentiment")
confusion_matrix(Y1_test,Y1_pred)
```

```python
print("Confusion Matrix of Neutral Sentiment")
confusion_matrix(Y2_test,Y2_pred)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy score of Sentiment_Positive",accuracy_score(Y1_test, Y1_pred)*100)
print("Accuracy score of Sentiment_Neutral",accuracy_score(Y2_test, Y2_pred)*100)
```

# Applying Random Forest

```python
from sklearn.model_selection import train_test_split
X_train, X_test,Y1_train,Y1_test, Y2_train,Y2_test = train_test_split(X, Y1,Y2, test_size=0.25, random_state=0)
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy score of Sentiment_Positive",accuracy_score(Y1_test, Y1_pred)*100)
print("Accuracy score of Sentiment_Neutral",accuracy_score(Y2_test, Y2_pred)*100)
```

# Applying K-nearest neighbor

```python
from sklearn.model_selection import train_test_split
X_train, X_test,Y1_train,Y1_test, Y2_train,Y2_test = train_test_split(X, Y1,Y2, test_size=0.25, random_state=72)
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 1)
classifier.fit(X_train, Y1_train)
Y1_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
print("Confusion Matrix of positive Sentiment")
confusion_matrix( Y1_test,Y1_pred)
```

```python
print("Confusion Matrix of Neutrl Sentiment")
confusion_matrix( Y2_test,Y2_pred)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy score of Sentiment_Positive",accuracy_score(Y1_test, Y1_pred)*100)
print("Accuracy score of Sentiment_Neutral",accuracy_score(Y2_test, Y2_pred)*100)
```
