```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
cproducts=pd.read_csv('cproducts.csv')
cproducts
```

```python
temp=pd.DataFrame(cproducts)
```

```python
temp
```

```python
temp.isnull().sum()
```

```python
sns.heatmap(temp.isnull(),cmap="Greens")
```

```python
temp.describe()
```

```python
temp.describe(include=['object'])
```

```python
temp.DOB
```

```python
dd='00'
mm='00'
yyyy='00'
print(dd+"-"+mm+"-"+yyyy)
```

```python
temp.DOB.fillna(value=dd+"-"+mm+"-"+yyyy,inplace=True)
```

```python
temp.DOB.isnull().sum()
```

```python
temp.Gender[temp.Gender=='female'].count()
```

```python
temp.Gender[temp.Gender=='male'].count()
```

```python
temp.Gender.fillna(value='male',inplace=True)
temp.State.fillna(value='no-state',inplace=True)
```

```python
temp.Gender.isnull().sum()
```

```python
sns.countplot(x='Gender',data=temp)
```

```python
plt.figure(figsize=(15,7))
chart=sns.countplot(x='State',data=temp,palette='Set1')
chart.set_xticklabels(chart.get_xticklabels(),rotation=90)
```
